```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('/Users/dhanushsiddharthapinjerla/Desktop/Elevate labs/task5/gender_submission.csv')  
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
for col in df.columns:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts(dropna=False))
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
survival_rate = df["Survived"].mean()
print(f"Survival rate: {survival_rate:.2%}")
```

```python
sns.scatterplot(x="PassengerId", y="Survived", data=df)
plt.title("Survival vs PassengerId")
plt.show()
```

```python
df.hist(figsize=(8,6), bins=15)
plt.suptitle("Histograms of numeric columns")
plt.show()
```

```python
for col in df.select_dtypes(include="number").columns:
    plt.figure(figsize=(6,2))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
if df.shape[1] > 1:  
    sns.scatterplot(x="PassengerId", y="Survived", data=df)
    plt.title("Scatterplot: Survived vs PassengerId")
    plt.show()
```
